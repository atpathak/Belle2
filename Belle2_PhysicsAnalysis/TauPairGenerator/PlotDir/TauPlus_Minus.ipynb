```python
import ROOT
from ROOT import Belle2, TH1F, TH2F, TCanvas, THistPainter, TPad, TFile
infile = TFile('../tau_Generator_tauegamma.root')
myTree = infile.Get("tree")
hist_tauPlus = ROOT.TH1F("hist_tauPlus", "tau Plus Decays", 100,0,50);
hist_tauMinus = ROOT.TH1F("hist_tauMinus", "tau Minus Decays", 100,5,55);
for entry in myTree:
    tauPlus = entry.tauPlusMCMode
    tauMinus = entry.tauMinusMCMode
    hist_tauPlus.Fill(tauPlus)
    hist_tauMinus.Fill(tauMinus)
```

```python
ROOT.gStyle.SetOptStat(0)
canvas = ROOT.TCanvas("canvas", "canvas", 1600, 1600)
hist_tauPlus.Scale(1./hist_tauPlus.Integral());
hist_tauPlus.GetXaxis().SetTitle("tau Plus decay modes")
hist_tauPlus.GetYaxis().SetTitle(" branching ratio")
hist_tauPlus.Draw("hbar01")
canvas.Draw()
```

```python
ROOT.gStyle.SetOptStat(0)
canvas = ROOT.TCanvas("canvas", "canvas", 1600, 1600)
hist_tauMinus.Scale(1./hist_tauMinus.Integral());
hist_tauMinus.GetXaxis().SetTitle("tau Minus decay modes")
hist_tauMinus.GetYaxis().SetTitle(" branching ratio")
hist_tauMinus.Draw("hbar01")
canvas.Draw()
```
